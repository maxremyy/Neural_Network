### **2. Model Creation**

#### 2.1 Data Normalization

- For optimal training, the **RGB values need to be normalized**.
- This involves **scaling down to values between 0 and 1**.
- The normalization is quite simple, as the values just need to be **divided by 255**.

```python
# Normalization of the RGB values of both datasets.

x_train = tf.keras.utils.normalize(x_train, axis = 1)
x_test = tf.keras.utils.normalize(x_test, axis = 1)
```

```python
# Checking if the values have been normalized.

x_train[0]
```

#### 2.2 Model Creation

- Next, we will **create, compile, and train the neural network**.
- A model is an object that **groups layers together** and can be **trained on data**.
- The simplest type of model is the **sequential model**, which is a **linear stack of layers**.



**Network Setup**
- With **.Sequential()**, the network is created.
- The **Con2D-Layer** is the first Convolutional Layer in the neural network. It looks at small pieces of an image and identifies patterns, like lines or circles. It does this by moving little windows (filters) over the image and noting important features. This helps the network learn to recognize objects and shapes - [Visualisation of a Conv2D-Layer](https://animatedai.github.io/) **Convolution is a process** that involves sliding a filter (or kernel) over an image and performing element-wise multiplication with the parts of the image it covers, then summing up those results into a single output pixel in a feature map. 
- The **MaxPooling2D-Layer** reduces the size of the input image or feature map by taking the largest value in each small patch of the image. This process helps to make the neural network more efficient and reduces the amount of computation needed, while still preserving important features.
- The **.Flatten() method** converts the multidimensional input data into a single long 1D vector.
- One **hidden layers** is created, with 64 neurons. 
- The **activation function** of the convolutional layers and the hidden layer is a **ReLU function** (Rectified Linear Unit). If the input values are negative, the ReLU function outputs a 0. If the input values are positive, the ReLU function outputs the input value.
- The **output layer** has 10 neurons, one for each item. The activation function of the output layer is a **softmax function**. This outputs values between 0 and 1.

```python
model = tf.keras.models.Sequential([
    
    tf.keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    
    tf.keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
        
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

**Compiling the Network**
- With **.compile()**, the network is prepared for training.
- The relevant parameters for us are **optimizer, loss, and metrics**.
- Optimizer: The optimization algorithm that **performs weight adjustments during training**. Since the learning rate is not further defined, it defaults to the default learning rate of 0.001.
- Loss: The loss function measures **how well the model performs on the training data**.
- Metrics: The metrics to be monitored during training to **evaluate the model's performance**.

```python
# Here, the model is prepared for training.

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

**Training the Network**
- With **.fit()**, the network is trained.
- During this process, the **weights between the neurons and the biases** are adjusted to improve the quality of the prediction.
- In each epoch, all **training data pass through the network once**, so the network sees each image 6 times. The model is trained in 6 epochs.
- The training data are divided into **batches**, each with 32 (default) images.
- After a batch has passed through the network, the **weights are adjusted**. This process repeats for all batches. In this case, 50,000 / 32 = 1,562.5.
- The **output indicates the size of the loss (error) and the accuracy for the training and test** datasets. These figures help us analyze the training later and draw conclusions about the model quality. We will need these details later.

```python
# The model is trained.

model_history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

```python
# The model parameters are displayed.

model.summary()
```

**Summary**
- In the network, **167,562 parameters (weights and biases)** were trained, which are composed as follows:
- **First Conv2D Layer**: 864 weights from the **input layer to the first Conv2D Layer** (3 * 3 * 3 * 32 = 864) plus 32 biases for each filter - **Total: 896**.
- **First MaxPooling Layer:** The output is a **3D array (30,30,32)**, where the last dimension represents the number of channels, in our case 32, one for each filter. This "cube of values" is reduced by the MaxPooling2D layer to a size of **15,15,32**. The number of filter outputs remains the same. These values (total = 7,200) are then passed on to the second Conv2D layer.
- **Second Conv2D Layer**: 18.432 weights from the **first MaxPooling layer to the second Conv2D Layer** (3 * 3 * 32 = 288 weights for each filter, times  64 filters = 18.432) plus 64 biases for each filter - **Total: 18.496**.
- **Second MaxPooling Layer:** The output is a **3D array (13,13,64)**, where the last dimension represents the number of channels, in our case 64, one for each filter. This "cube of values" is reduced by the MaxPooling2D layer to a size of **6,6,64**. The number of filter outputs remains the same. These values (total = 2,304) are then passed on to the next step.
- **Flatten** converts the output of 6,6,64 into a **1D vector with a length of 2,304**.
- **The first dense hidden layer** consists of 64 neurons. The 1D vector with 2,304 values is fully connected to these 64 neurons, creating 147,456 weights. Adding 64 biases to these weights results in a total of 147,520 parameters.
- **The output layer consists** of 10 neurons. The 64 neurons from the preceding dense layer are fully connected to these 10 neurons, creating 640 weights. Adding 10 biases to these weights results in a total of 650 parameters.

**Excursion into Overfitting**

- To check if the model is not overfitting, we look at the **training history**.
- Overfitting: The model **memorizes the specific patterns** of the training data but fails to acquire the ability **to recognize general patterns**.
- Below, the **performance of the model after each epoch** of the training process is graphically displayed.
- If the model were **overfitting**, then **the values for the val_accuracy would not increase**.
- That is not the case here; therefore, **the model is not overfitted**.

### **3. Model Evaluation**

#### 3.1 Model Assessment

- Now, the **network's performance is evaluated**.
- How well does the network **classify new images**? 

```python
y_predictions = model.predict(x_test)
```

```python
y_predictions[0]
```

```python
y_predictions = [np.argmax(arr) for arr in y_predictions]
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_predictions))
```

**Summary**

- Fundamentally, the network can make **two mistakes when classifying**, for example, a Shirt is classified as another piece of clothing, or another piece of clothing is classified as a Shirt. These errors are also called **"False positives"** (fälschlicherweise positiv) and **"False negatives"** (fälschlicherweise negativ).
- For each piece of clothing, one can create a **confusion matrix** and calculate the following values:
- **Precision** = (true positive / (true positive + false positive)) -> **How many of the positively predicted cases are actually positive**?
- **Recall** = (true positive / (true positive + false negative)) -> **How many of the truly positive cases were also identified as such**?
- True positives: Model predicts Shirt; in reality it is a Shirt.
- True negatives: Model predicts not a Shirt; in reality it is not a Shirt.
- False positives: Model predicts Shirt; in reality it is not a default Shirt.
- False negatives: Model predicts no Shirt; in reality it is a Shirt.
- The **accuracy indicates how many of the images were correctly classified**. Calculation: Number of correct predictions / Total number of predictions.
- Reminder: The numbers are assosited with the clothing as followed: **0 airplane 1 autombile, 2 bird, 3 cat, 4 deer, 5 dog, 6 frog, 7 horse, 8 ship, 9 truck**
- It can be seen that the network, for example, classifies approximately 75% of the trucks (label 9) as trucks (Recall). Thus, 25% of the trucks are classified as another item.
- In total, approximately **65% of all images are correctly classified**.

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predictions)
```

```python
cm
```

```python
# Creating a heatmap of the confusion matrix with matplotlib

fig, ax = plt.subplots(figsize=(6, 6))
cax = ax.matshow(cm, cmap='Blues')

for (i, j), val in np.ndenumerate(cm):
    ax.text(j, i, val, ha='center', va='center', color='gray')

plt.colorbar(cax)
plt.ylabel('Truth')
plt.xlabel('Predictions')
plt.title('Confusion Matrix')
plt.show()
```
